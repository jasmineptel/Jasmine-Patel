# Machine Learning Basics

## Artificial Intelligence (AI)

Artificial Intelligence is a wide range of technologies that aim to simulate human intelligence, including natural language processing (NLP), computer vision, robotics, and machine learning.

## Machine Learning (ML)

The use and development of computer systems that are able to learn and adapt without following explicit instructions, by using algorithms and statistical models to analyze and draw inferences from patterns in data. There are several types of machine learning including supervised ML, unsupervised ML, deep learning, and reinforcement Learning.

## Parametric Models
Parametric models have a fixed, finite number of parameters and assumes that the data follows a specific distribution (like a normal distribution). Examples include Linear regression, Logistic regression, Naive Bayes classifier

## Non-Parametric Models
Non-Parametric models can have an infinite number of parameters, effectively allowing them to adapt to more complex data shapes. Unlike parametric models, they do not make assumptions regarding distribution of data. Examples include K-Nearest Neighbors (KNN), Decision trees, Support Vector Machines (SVM)

### K-Nearest Neighbors

Under the hood explanation of KNN

```python
def knnclassify(test_data, training_data, training_labels, K=1):
  y_pred = []

  for i in range(len(test_data)):
    distances = []
    for j in range(len(training_data)):
      dist = np.sqrt(np.sum((training_data[j] - test_data[i]) ** 2)) #Uses Euclidean distance between test points and training points
      distances.append(dist)

    sort_dists = np.argsort(distances) #Gets indexes of sorted distances
    labels = training_labels[sort_dists[:K]] #Gets K nearest neighbors

    class_labels = Counter(labels) #Counts labels of the K nearest neighbors
    y_pred.append(class_labels.most_common(1)[0][0]) #Finds the majority vote

  return y_pred
```

## Unsupervised Learning

Under the hood explanation for K Means

```python
def kmeans_clustering(all_vals, K, max_iter = 100, tol=pow(10, -3)):
  #Assigning clusters and start iterations
  idxs = np.random.choice(len(all_vals), K, replace=False)
  centroids = all_vals[idxs, :]
  it = 0

  #Keeps track of error per iteration
  all_sse = []

  #Keeps track of most recent cluster assignments
  ending_clusters = []

  while (it < max_iter):
    clusters = []

    #Use this for calculating sse for iteration i
    dist_for_sse = []

    for i in range(len(all_vals)):
        #Calculate the distance from the data point to all centroids
        find_dist = []

        #Compute distances of each centroid from a given data point
        for j in range(len(centroids)):
          distances = np.linalg.norm(all_vals[i] - centroids[j])
          find_dist.append(distances)

        #Takes index of the minimum value and assigns centroid
        closest_centroid = np.argmin(find_dist)

        #Append minimum distance to calc sse
        dist_for_sse.append(min(find_dist))

        #Assign the data point to the nearest centroid
        clusters.append(closest_centroid)


    #Updates clusters
    update_centroids = []

    #Updates clusters by taking the mean of all the columns of each feature
    #to re-assign clusters
    for i in range(K):
      cluster_points = []
      for k in range(len(clusters)):
        if clusters[k] == i:
          cluster_points.append(all_vals[k])

      #Accounts for the case where certain clusters may be empty
      #If so reassigned the centroids to random data points
      if (len(cluster_points) == 0):
          update_centroids.append(all_vals[np.random.choice(len(all_vals))])
      else:
          take_mean = np.mean(cluster_points, axis=0)
          update_centroids.append(take_mean)

    #Add sse for this iteration
    all_sse.append(np.mean(dist_for_sse))

    #Check for convergence using the given SSE condition
    if len(all_sse) > 1 and np.absolute(all_sse[it] - all_sse[it-1])/ all_sse[it-1] <= tol:
      break

    #returning centroids and assignments for this iteration
    centroids = list(update_centroids)
    ending_clusters = list(clusters)

    it += 1

  return clusters, centroids, all_sse, it
```
